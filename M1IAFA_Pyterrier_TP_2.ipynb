## TP 2. Application de la RI pour la liaison référentielle 

Une application des méthodes de RI est la liaison référentielle. Cette tâche consiste à trouver le lien entre une portion de texte et la page Wikipédia qui le représentent le mieux. Dans ce cas d’application :
* Comme requêtes nous allons utiliser chaque mot dans une phrase (ou chaque mot important).
* Le corpus sont les pages Wikipédia. Donc, chaque page Wikipédia est un document à retrouver.
* La liste de résultats sont les candidats pour la liaison référentielle.

Considère par exemple la chaîne de caractères « ***Thomas and Mario are strikers playing in Munich*** ». Il existe trois sous-chaînes (***Thomas***, ***Mario*** et ***Munich***) ou mots importants qui sont susceptibles d'être identifiées dans une base de connaissances telle que Wikipédia. Une recherche de ces mots-clés dans les pages Wikipédia produirait de nombreux concepts candidats pour ces sous-chaînes. Donc un bon système doit retrouver comme premier résultats les liens aux pages Wikipédia : [Thomas Müller](https://en.wikipedia.org/wiki/Thomas_M%C3%BCller), [Mario Gómez](https://en.wikipedia.org/wiki/Mario_G%C3%B3mez) (ou Mario Götze) et FC [Bayern Munich](https://en.wikipedia.org/wiki/FC_Bayern_Munich). Cette tâche est particulièrement complexe et les méthodes de recherche d’information sont particulièrement utiles pour la résoudre.

> Plusieurs méthodes de liaison référentielle (Entity Linking en anglais) sont disponibles sur internet. Les plus connus sont : 

>	* Spotlight → https://demo.dbpedia-spotlight.org/
>	* Babelfy → http://babelfy.org/

```
models = ["DFIZ", "DPH", "TF_IDF", "BM25", "DFIC", "BB2","DFR_BM25"]

nb_premiers = [10, 20, 50]
for nb in nb_premiers : 
  for model in models :
    x = pt.BatchRetrieve(indexref2, wmodel=model, metadata=["docno","title","url"]).search("Thomas and Mario are strikers playing in Munich").head(nb)
    if "Thomas Mller" in x.title.values :
      print("Thomas Müller trouvé par ", model,"dans les ",nb,"premiers")

    if "Mario Gmez" in x.title.values :
      print("Mario Gomez trouvé par ", model,"dans les ",nb,"premiers")

    if "Mario Gotze" in x.title.values :
      print("Mario Gtze trouvé par", model,"dans les ",nb,"premiers")

    if "FC Bayern Munich" in x.title.values :
      print("FC Bayern Munich trouvé par", model,"dans les ",nb,"premiers")
```

### 3) Utilisation de termes
Une stratégie consiste à utiliser individuellement chaque terme de la requête pour voir si vous arrivez à trouver les pages d’intérêt. Un code exemple est donné ci-dessous, cependant il n'est pas suffisant pour faire la tâche. Une fois fini votre code, réalisez la même vérification de résultats que dans le point précédent.


```
query = "Thomas and Mario are strikers playing in Munich"
termToSearch = [x for x in query.split() if x[0]==x[0].upper()]
models = ["DFIZ", "DPH", "TF_IDF", "BM25", "DFIC", "BB2", "DFR_BM25"]
nb_premiers = [10, 20, 50]

for nb in nb_premiers :
  for model in models:
    for term, page in zip(termToSearch,['Thomas Müller','Mario Götze', 'FC Bayern Munich']):
      # Recherche du terme dans l'index
      dfres = pt.BatchRetrieve(indexref2, wmodel=model, metadata=["docno","title","url"]).search(term)[['title','score']].head(nb)
      print("Modèle : ", model, "\n", dfres)
      print("Position de la page attendue dans les résultats :\n",dfres[dfres['title']==page],"\n\n\n")
```

Le fichier anchor_text_en.tll  contient les informations de la Wikipédia en anglais et pas de la Wikipédia simple, donc certain titres n’ont pas de correspondance.

```
def search_wikipedia_pages(paragraph):
    # Initialisation de l'index de recherche de Wikipédia
    index = pt.IndexFactory.of("terrier_wikipedia")

    # Tokenisation du paragraphe en mots
    words = paragraph.split()

    # Initialisation d'un dictionnaire pour stocker les résultats de recherche
    results = {}

    # Parcours de chaque mot dans le paragraphe
    for word in words:
        # Recherche des pages Wikipédia pour le mot courant
        query = pt.BatchRetrieve(index, wmodel="BM25", metadata=["docno","title","url"]).setControl('decorate', 'on')
        res = query.search(word)[['title','url','score']]

        # Ajout des résultats de recherche dans le dictionnaire
        results[word] = res

    # Transformation du dictionnaire de résultats en un dataframe
    df_results = pd.concat([v.assign(word=k) for k,v in results.items()])

    # Trier les résultats par mot et score décroissant
    df_results = df_results.sort_values(['word','score'], ascending=[True,False])

    # Grouper les résultats par mot et renvoyer une liste de pages pour chaque mot
    results_list = []
    for word in words:
        res = df_results[df_results['word'] == word][['title','url']].to_dict(orient='records')
        results_list.append(res)

    return results_list
```
